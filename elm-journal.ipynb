## 1. Topic: Exploring the String Module

- [ ]  String.words, String.join, String.filter, endsWith

I was working on a little Exercise called 'Bob':

- Bob is a lackadaisical teenager. In conversation, his responses are very limited.

- Bob answers 'Sure.' if you ask him a question, such as "How are you?".

- He answers 'Whoa, chill out!' if you YELL AT HIM (in all capitals).

- He answers 'Calm down, I know what I'm doing!' if you yell a question at him.

- He says 'Fine. Be that way!' if you address him without actually saying anything.

- He answers 'Whatever.' to anything else.

- Bob's conversational partner is a purist when it comes to written communication and always follows normal rules regarding sentence punctuation in English.

### 1.1. Yelling on Bob: String.all and empty Strings
- I wanted to check if all characters were uppercase. So at first I went with: 'String.all Char.isUpper' - to check if all characters are uppercase.
- I then realized from the test suite that a 'shouting' String to Bob could contain whitespaces (of course: words!), exclamation marks, special characters and of course question marks (questions!). So `Char isUpper` on its own won't do the job.
- So I went with 'String.filter Char.isAlpha' to filter out the alphabetic characters and check if they are all uppercase.

I then got over something unexpected:

> String.all returns true if you pass an empty String to it. [--> Insights on why this is on Discourse](https://discourse.elm-lang.org/t/empty-strings-with-string-all/5062/4)


Verification example 1.1.1

````elm
import Html

main =
    if String.all Char.isAlpha "" then
        Html.text "-- Verify: The empty string qualifies as 'All characters are letters'"
    else
        Html.text ""
        
-- compile-code
````

Verification example 1.1.2

````elm
import Html

main =
    if String.all Char.isUpper "" then
        Html.text "-- Verify: The empty string qualifies as 'All characters are uppercase'"
    else
        Html.text ""
        
-- compile-code
````

- Given an empty String, all charactes qualify as alphabetic

### 1.2. Addressing Bob Silent (checking for whitespace)

- I was looking for a way in the `String` and `Char` module on how to test a String if it has whitespaces in it. The Testing suite for Bob was feeding different whitespace characters into the tests.

```
test "alternate silences" <|
            \() ->
                Expect.equal
                    "Fine. Be that way!"
                    (Bob.hey "\t  \n  \t   ")
```

```
test "other whitespace" <|
            \() ->
                Expect.equal
                    "Fine. Be that way!"
                    (Bob.hey "\n\u{000D} \t")
```                    


- So since I found nothing in the String Module, the closest solution for me was just using a Regex:

````elm
import Html
import Regex

isAddressingSilent remark =
    case Regex.fromString "^\\s*$" of
        Just whiteSpaceRegex ->
            Regex.contains whiteSpaceRegex remark

        Nothing ->
            False
            
main =
    let
        adressSilent =
            isAddressingSilent "\n\u{000D} \t"
    in
    if adressSilent then
        Html.text "Fine. Be that way !"
    else
        Html.text "Whatever."
-- compile-code
````

**Reflection:**

**However**, in reflection, a better way to pass the whitespace tests is to just trim the remark and check if the result was empty:

````elm
import Html

isAddressingSilent remark =
    String.trim remark |> String.isEmpty
            
main =
    let
        adressSilent =
            isAddressingSilent "\n\u{000D} \t"
    in
    if adressSilent then
        Html.text "Fine. Be that way !"
    else
        Html.text "Whatever."
        
-- compile-code
````

## 11. Message naming conventions

- https://www.youtube.com/watch?v=w6OVDBqergc
- What happened ? -> DidSomething (Clicked, Hovered, GotData)
- Who or what ? Someone / Something
- Optionally: Where / How ? SomewhereAndSomeHow
- -> SomeoneDidSomethingSomewhereAndSomeHow
- Being more specific in the first place is better than being abstract (Start specific, get abstract) - the nescessary abstractions will reveal themselves.

Examples:

- UserMovedMouse
- UserClickedSignInButton
- ServerRespondedWithAccountData
- WorkerProcessedJob
- BillingServiceCompletedSubscriptionAdd
